```python
import pickle as pkl
```

```python
data = open("important.pkl", "rb")
```

```python
dict1 = pkl.load(data)
```

```python
import matplotlib.pyplot as plt
```

```python
dict1.keys()
```

```python
plt.scatter(dict1['original'], dict1['fitted'])
plt.xlabel("Before isotonic prediction")
plt.ylabel("After isotonic prediction")
```

```python
file = open("observed.pkl", "rb")
obcdict = pkl.load(file)
```

```python
obcdict['observed'].shape
```

```python
dict1['fitted'].shape
```

```python
import scipy.stats
import numpy as np
plt.scatter(obcdict['observed'], dict1['fitted'], alpha=0.1, s=1)
plt.xlabel("Observed")
plt.ylabel("After isotonic predicted")
plt.title("Pearsonr = "+str(np.round(scipy.stats.pearsonr(obcdict['observed'][~np.isnan(dict1['fitted'])], dict1['fitted'][~np.isnan(dict1['fitted'])])[0],2)))
```

```python
plt.scatter(obcdict['observed'], dict1['original'],  alpha=0.1, s=1)
plt.xlabel("Observed")
plt.ylabel("Before isotonic predicted")
plt.title("Pearsonr = "+str(np.round(scipy.stats.pearsonr(obcdict['observed'][~np.isnan(dict1['original'])], dict1['original'][~np.isnan(dict1['original'])])[0],2)))
```
